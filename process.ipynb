```python
import pandas as pd
DATA_PATH = '../data/raw/EMI_dataset.csv'

df = pd.read_csv("emi_prediction_dataset.csv")

print(f'Shape : {df.shape[0]:,} rows × {df.shape[1]} columns')
df.head()
```

```python
print(f"Total Rows    : {df.shape[0]:,}")
print(f"Total Columns : {df.shape[1]}")
print(f"\nMissing Values per Column:")
print(df.isnull().sum())
```

```python
df.describe().T
```

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100

missing_report = pd.DataFrame({
    'missing_count' : missing,
    'missing_%'     : missing_pct.round(2)
})

print(missing_report[missing_report['missing_count'] > 0])
```

```python
print("Credit score - invalid values:")
print(f"  Values below 300 : {(df['credit_score'] < 300).sum()}")
print(f"  Values above 850 : {(df['credit_score'] > 850).sum()}")
print(f"  Zero values      : {(df['credit_score'] == 0).sum()}")
print(f"  Missing values   : {df['credit_score'].isnull().sum()}")
```

```python
import numpy as np
# Replace 0s and values above 850 with NaN, then fill with median
df['credit_score'] = df['credit_score'].replace(0, np.nan)
df.loc[df['credit_score'] > 850, 'credit_score'] = np.nan

print(f"Credit score NaN after cleaning : {df['credit_score'].isnull().sum()}")
print(f"New min : {df['credit_score'].min()}")
print(f"New max : {df['credit_score'].max()}")
```

```python
# Check which columns are object type (strings)
print("Column dtypes:")
print(df.dtypes)
```

```python
# These columns look numeric but are stored as strings
numeric_cols = ['age', 'monthly_salary']

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    print(f"{col} → dtype: {df[col].dtype}, nulls: {df[col].isnull().sum()}")
```

```python
df['bank_balance'] = pd.to_numeric(df['bank_balance'], errors='coerce')
df['bank_balance'] = df['bank_balance'].fillna(df['bank_balance'].median())
print(f"bank_balance → dtype: {df['bank_balance'].dtype}, nulls: {df['bank_balance'].isnull().sum()}")
```

```python
# age → fill with median (small count, 3 rows)
df['age'] = df['age'].fillna(df['age'].median())

# monthly_salary → fill with median (financial column, skewed)
df['monthly_salary'] = df['monthly_salary'].fillna(df['monthly_salary'].median())

print(f"age nulls remaining           : {df['age'].isnull().sum()}")
print(f"monthly_salary nulls remaining: {df['monthly_salary'].isnull().sum()}")
```

```python
df['credit_score']  = df['credit_score'].fillna(df['credit_score'].median())
df['monthly_rent']  = df['monthly_rent'].fillna(df['monthly_rent'].median())
df['emergency_fund']= df['emergency_fund'].fillna(df['emergency_fund'].median())
df['education']     = df['education'].fillna(df['education'].mode()[0])

print("Missing values remaining after all fills:")
remaining = df.isnull().sum()
print(remaining[remaining > 0])
print(f"\nTotal missing : {df.isnull().sum().sum()}")
```

```python
cat_cols = ['gender', 'marital_status', 'education', 'employment_type',
            'company_type', 'house_type', 'existing_loans', 'emi_scenario', 'emi_eligibility']

for col in cat_cols:
    print(f"{col}: {df[col].unique().tolist()}")
```

```python
# Standardize gender column
df['gender'] = df['gender'].str.strip().str.upper()
df['gender'] = df['gender'].replace({
    'FEMALE' : 'Female',
    'MALE'   : 'Male',
    'F'      : 'Female',
    'M'      : 'Male'
})

print(f"Gender unique values after cleaning: {df['gender'].unique().tolist()}")
print(df['gender'].value_counts())
```

```python
# Label encoding using exact values we just confirmed

df['gender'] = df['gender'].map({'Female': 0, 'Male': 1})

df['marital_status'] = df['marital_status'].map({'Single': 0, 'Married': 1})

df['education'] = df['education'].map({
    'High School'   : 0,
    'Graduate'      : 1,
    'Post Graduate' : 2,
    'Professional'  : 3
})

df['employment_type'] = df['employment_type'].map({
    'Private'      : 0,
    'Government'   : 1,
    'Self-employed': 2
})

df['company_type'] = df['company_type'].map({
    'Small'       : 0,
    'Startup'     : 1,
    'Mid-size'    : 2,
    'Large Indian': 3,
    'MNC'         : 4
})

df['house_type'] = df['house_type'].map({
    'Rented': 0,
    'Family': 1,
    'Own'   : 2
})

df['existing_loans'] = df['existing_loans'].map({'No': 0, 'Yes': 1})

df['emi_scenario'] = df['emi_scenario'].map({
    'E-commerce Shopping EMI': 0,
    'Home Appliances EMI'    : 1,
    'Vehicle EMI'            : 2,
    'Personal Loan EMI'      : 3,
    'Education EMI'          : 4
})

df['emi_eligibility'] = df['emi_eligibility'].map({
    'Not_Eligible': 0,
    'High_Risk'   : 1,
    'Eligible'    : 2
})

print("Encoding done. Checking for NaNs introduced by mapping:")
print(df[cat_cols].isnull().sum())
```

```python
print("Encoding done. Checking for NaNs introduced by mapping:")
print(df[cat_cols].isnull().sum())
```

```python
print("\nDtypes after encoding:")
print(df.dtypes)
```

```python
before = len(df)
df = df.drop_duplicates().reset_index(drop=True)
print(f"Rows before : {before:,}")
print(f"Rows after  : {len(df):,}")
print(f"Duplicates removed : {before - len(df):,}")
```

```python
TARGET_CLASS = 'emi_eligibility'
TARGET_REG   = 'max_monthly_emi'

X = df.drop(columns=[TARGET_CLASS, TARGET_REG])
y_class = df[TARGET_CLASS]
y_reg   = df[TARGET_REG]

print(f"Features shape : {X.shape}")
print(f"Classification target classes : {sorted(y_class.unique())}")
print(f"Regression target range : {y_reg.min():.0f} – {y_reg.max():.0f} INR")
print(f"\nFeature columns:")
for i, col in enumerate(X.columns):
    print(f"  {i+1:>2}. {col}")
```

```python
from sklearn.model_selection import train_test_split

# Step 1 — Split out test set (15%)
X_temp, X_test, yc_temp, yc_test, yr_temp, yr_test = train_test_split(
    X, y_class, y_reg,
    test_size=0.15,
    random_state=42,
    stratify=y_class
)

# Step 2 — Split remaining into train (70%) and validation (15%)
X_train, X_val, yc_train, yc_val, yr_train, yr_val = train_test_split(
    X_temp, yc_temp, yr_temp,
    test_size=0.176,
    random_state=42,
    stratify=yc_temp
)

print(f"Train      : {len(X_train):,} rows")
print(f"Validation : {len(X_val):,} rows")
print(f"Test       : {len(X_test):,} rows")
print(f"Total      : {len(X_train)+len(X_val)+len(X_test):,} rows")
```

```python
import os
os.makedirs('../data/processed', exist_ok=True)

# Save clean full dataset
df.to_csv('../data/processed/df_clean.csv', index=False)

# Save splits
X_train.to_csv('../data/processed/X_train.csv', index=False)
X_val.to_csv('../data/processed/X_val.csv',     index=False)
X_test.to_csv('../data/processed/X_test.csv',   index=False)

yc_train.to_csv('../data/processed/yc_train.csv', index=False)
yc_val.to_csv('../data/processed/yc_val.csv',     index=False)
yc_test.to_csv('../data/processed/yc_test.csv',   index=False)

yr_train.to_csv('../data/processed/yr_train.csv', index=False)
yr_val.to_csv('../data/processed/yr_val.csv',     index=False)
yr_test.to_csv('../data/processed/yr_test.csv',   index=False)

# Save feature column list
import json
with open('../data/processed/feature_cols.json', 'w') as f:
    json.dump(X.columns.tolist(), f)

print("All files saved to ../data/processed/")
print(os.listdir('../data/processed/'))
```
